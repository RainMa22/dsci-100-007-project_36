## Introduction

Pulsars are highly magnetized, rotating neutron stars that emit a beam of electromagnetic radiation, similarly to a lighthouse, detectable from Earth only when aligned with our line of sight. Studying pulsars provides an opportunity to study matter undergoing extreme conditions, and stellar evolution. Pulsars also help astronomers study gravitational waves, an important concept for understanding spacetime, and the history of our universe. However, the rarity of pulsars and the significant noise in radio signal data pose a significant challenge to their detection and classification.

This project seeks to cut through the cosmic static to accurately identify these enigmatic objects. We will harness the power of data science, utilizing the HTRU2 dataset from the UC Irvine Machine Learning Repository, to construct a classification model capable of discerning pulsars from a cacophony of radio frequency interference. The HTRU2 dataset is a compendium of labeled data, detailing characteristics of integrated pulse profiles and DM-SNR curves from observed celestial sources.

Through methodical analysis and machine learning, specifically K-Nearest Neighbors (KNN), we aim to predict whether a radio signal candidate is a pulsar. This endeavor is not just academic; successfully classifying pulsars could streamline the process of their discovery, enhancing our capacity for astronomical research and potentially revealing new cosmic truths.

## Methods and Results

### Data Acquisition and Preprocessing
Our analysis began with the acquisition of the HTRU2 dataset from the UC Irvine Machine Learning Repository. The data was downloaded and loaded into an R environment for processing. Initial preprocessing steps included:
 - **Loading the Libraries:** The libraries required for this project include the tidyverse package which helps with tidying, wrangling, and visualizing the data via ggplot. We also use the tidymodels package which allows us to create the models for this project.
 - **Unzipping the Dataset:** Firstly, we will download the dataset by calling download.file(); Then, utilizing R's built-in unzip function, the zipped file containing the dataset was extracted.
 - **Data Loading:** The dataset was then read into a dataframe using R’s read_csv function, with the columns named appropriately to reflect the data they represent: Profile mean, Profile standard deviation, Profile skewness, Profile kurtosis, DM-SNR mean, DM-SNR standard deviation, DM-SNR skewness, and DM-SNR kurtosis, along with the target variable 'class'.
 - **Data wrangling:** We conducted a preliminary check for missing values and inconsistencies in the dataset. Using the initial_split() function, and by a stratum(on target variable ‘class’), we have partitioned the original data to create a 75:25 split between the training and testing sets, ensuring that both sets were representative of the overall distribution of the data. The data were surprisingly clean overall, so no further wrangling was introduced.
### Exploratory Data Analysis (EDA)
We embarked on an exploratory analysis to better understand the data and its structure:
 - **Visualizations:** plots were generated to observe the distributions of the features. A particularly telling graph was the pulsars’ mean signal profile against the mean DM-SNR Curve, which suggested a binary-like distribution, bolstering our confidence in the feasibility of classification.

### Model Building and Validation
For the prediction model, we selected the K-Nearest Neighbors (KNN) algorithm due to its efficacy in binary classification problems:
 - **Model Training:** Using the caret package in R, we trained a KNN classifier on the training set, optimizing the choice of 'k' through cross-validation to balance bias and variance effectively.
Feature Selection: All eight features were considered for the initial model, with the importance of each feature evaluated based on its contribution to classification accuracy.
Model Evaluation: We assessed the model's performance on the test set using metrics such as accuracy, precision, recall. A recall rate above 75% was set as the benchmark for successful classification.
### Results
The KNN model exhibited promising classification performance, with the following key metrics:
 - **Accuracy:** 97.5%, indicating the overall proportion of correct classifications made by the model.
 - **Recall:** 81.2%, fulfilling our objective of a recall rate above 75%, crucial for identifying the majority of true pulsars in the dataset.
 - **Precision:** 93.5% precision.

 - **Confusion Matrix:** This provided a clear picture of the model's true positive and negative rates, as well as instances of false positives and negatives.

During the search for the best number of neighbors for our KNN model, we discovered that K=7 and K=8 yields the same accuracy to the model, and we eventually settle on K=7, as an odd number of neighbors is preferred as there would be a tie-breaker.
The model achieved an accuracy of 97.8% and a recall rate of 81.2%, indicating a strong predictive performance, particularly in the context of recall, which is crucial for minimizing false negatives in pulsar identification. The precision of 93.5% further attest to the model's reliability.
### Reflection on Expectations
The high recall rate surpassed our initial benchmark expectation of 75%, which underscores the model's potential utility in astronomical research. This outcome resonates with our hypothesis that machine learning could significantly enhance the pulsar detection process against the backdrop of noisy data.
Impact of Findings
The successful classification of pulsars has profound implications for astrophysics. It can potentially streamline the discovery of pulsars, contributing to our understanding of the universe's fundamental physics. Moreover, the model's efficiency could expedite the identification process in large datasets, thereby freeing up valuable time for astronomers to focus on in-depth analysis.
### Future Research
The promising results of our project pave the way for several research opportunities:
Comparison with Other Models: Future work could compare the effectiveness of other algorithms, like support vector machines or neural networks, providing a comprehensive view of machine learning applications in pulsar classification.
Real-Time Detection: Integrating the classification model with telescope data feeds to develop a real-time detection system could revolutionize how we monitor and discover celestial bodies.

### Limitations
While the model is robust, there are limitations to consider. The classification accuracy is subject to the quality and range of data provided. Incomplete or biased datasets could lead to underperformance or misclassifications.
## Conclusion
In conclusion, our findings indicate that models like KNN can effectively classify pulsars from radio signal data, demonstrating significant potential to aid astronomical research. As we continue to refine our methods and expand our datasets, the horizon of understanding these enigmatic stars continues to broaden.
